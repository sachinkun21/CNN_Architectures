#### Why does AlexNet achieves better results than LeNet?

1. **Relu activation function for faster Training.**

Relu function: f (x) = max (0, x)

![alex1](img/alex512.png)


![alex1](img/alex612.png)

2. **Standardization ( Local Response Normalization )**


![alex1](img/alex3.jpg)


3. **Dropout**



![alex1](img/alex4.jpg)


4. **Enhanced Data ( Data Augmentation )**







```
import tensorflow as tf

class AlexNet(tf.keras.Model):
  def __init__(self, act_fun = 'relu', shape=(32,32,3), pred_shape = 10, padding = 'valid'):
    super(AlexNet,self).__init__()
    self.inp   = tf.keras.Input(shape=shape)
    
    self.conv1 = tf.keras.layers.Conv2D(filters = 96,kernel_size = (11,11),activation=act_fun, strides=(4,4), padding='valid')
    self.pool1 = tf.keras.layers.MaxPooling2D(pool_size=(3,3), strides=(2,2), padding = padding)
    self.BN_1  = tf.keras.layers.BatchNormalization()
    
    # 2nd Convolutional Layer
    self.conv2 = tf.keras.layers.Conv2D(filters=256, kernel_size=(5,5), strides=(1,1), padding='valid', activation=act_fun)
    self.pool2 = tf.keras.layers.MaxPooling2D(pool_size=(3, 3) ,strides=(2,2), padding = padding)
    self.BN_2  = tf.keras.layers.BatchNormalization()
    
    # 3rd Convolutional Layer
    self.conv3 = tf.keras.layers.Conv2D(filters = 384, kernel_size = (3,3), padding = padding, activation = act_fun)
    self.BN_3  = tf.keras.layers.BatchNormalization()
    # 4th Convolutional Layer
    self.conv4 = tf.keras.layers.Conv2D(filters = 384, kernel_size = (3,3), padding = padding, activation = act_fun)
    self.BN_4  = tf.keras.layers.BatchNormalization()
    # 5th Convolutional Layer
    self.conv5 = tf.keras.layers.Conv2D(filters = 256, kernel_size = (3,3), padding = padding, activation = act_fun)
    self.BN_5  = tf.keras.layers.BatchNormalization()
    
    # 3rd Pooling layer
    self.pool3 = tf.keras.layers.MaxPooling2D(pool_size = (3,3), strides = (2,2,), padding = padding)

    # Flatten
    self.flat = tf.keras.layers.Flatten()
    self.drop1= tf.keras.layers.Dropout(0.4)

    # 1st Fully Connected Layer
    self.fc1 = tf.keras.layers.Dense(4096, activation = act_fun)
    self.drop2= tf.keras.layers.Dropout(0.4)
    
    # 1st Fully Connected Layer
    self.fc2 = tf.keras.layers.Dense(4096, activation = act_fun)
    
    
    # Prediction layer
    self.pred = tf.keras.layers.Dense(pred_shape, activation = 'softmax')


  def call(self,inp):
    x = self.conv1(inp)
    x = self.pool1(x)
    x = self.BN_1(x)

    x = self.conv2(x)
    x = self.pool2(x)
    x = self.BN_2(x)

    x = self.conv3(x)
    x = self.BN_3(x)
    x = self.conv4(x)
    x = self.BN_4(x)
    x = self.conv5(x)
    x = self.BN_5(x)

    x = self.pool3(x)

    x = self.flat(x)
    x = self.drop1(x)

    x = self.fc1(x)
    x = self.drop2(x)

    x = x = self.fc2(x)
    
    return self.pred(x)


#AlexNet = AlexNet(act_fun = 'relu', shape=x.shape[1:], pred_shape = y.shape[1:][0])
```

### AlexNet on Cifar10

Note: Since image size is 28X28, changing padding to 'same' will help preserve the dimensions

```
import keras
from keras.datasets import cifar10
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dense, Flatten
from keras.models import Sequential

# Loading the dataset and perform splitting
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print(x_train.shape,y_train.shape)

# Peforming reshaping operation
x_train = x_train.reshape(x_train.shape[0], 32, 32, 3)
x_test = x_test.reshape(x_test.shape[0], 32, 32, 3)

print(x_train.shape,y_train.shape)

# Normalization
x_train = x_train / 255
x_test = x_test / 255

# One Hot Encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
print(x_train.shape,y_train.shape, x_test.shape,y_test.shape)
```

```
x_train.shape[1:]
```

```
alexnet_cifar = AlexNet(act_fun = 'relu', shape=x_train.shape[1:], pred_shape = y_train.shape[1:][0], padding = 'same')

batch_size = 64

alexnet_cifar.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse',metrics = ['accuracy'])
```

```
alexnet_cifar.fit( x_train, y_train, batch_size=batch_size, epochs=10, validation_data=(x_test,y_test), shuffle=True,  verbose=1)
alexnet_cifar.summary()
```
